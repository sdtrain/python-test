I had to modify the existing mean_absolute_error function of scikit-learn, so that I could calulate the mean_absolute_percentage_error (MAPE) as required by this task.

```python
# https://github.com/scikit-learn/scikit-learn/blob/7b136e9/sklearn/metrics/regression.py#L112

from sklearn.utils import check_array, check_consistent_length
from sklearn.externals.six import string_types

def _check_reg_targets(y_true, y_pred, multioutput):
    check_consistent_length(y_true, y_pred)
    y_true = check_array(y_true, ensure_2d=False)
    y_pred = check_array(y_pred, ensure_2d=False)

    if y_true.ndim == 1:
        y_true = y_true.reshape((-1, 1))

    if y_pred.ndim == 1:
        y_pred = y_pred.reshape((-1, 1))

    if y_true.shape[1] != y_pred.shape[1]:
        raise ValueError("y_true and y_pred have different number of output "
                         "({0}!={1})".format(y_true.shape[1], y_pred.shape[1]))

    n_outputs = y_true.shape[1]
    allowed_multioutput_str = ('raw_values', 'uniform_average',
                               'variance_weighted')
    if isinstance(multioutput, string_types):
        if multioutput not in allowed_multioutput_str:
            raise ValueError("Allowed 'multioutput' string values are {}. "
                             "You provided multioutput={!r}".format(
                                 allowed_multioutput_str,
                                 multioutput))
    elif multioutput is not None:
        multioutput = check_array(multioutput, ensure_2d=False)
        if n_outputs == 1:
            raise ValueError("Custom weights are useful only in "
                             "multi-output cases.")
        elif n_outputs != len(multioutput):
            raise ValueError(("There must be equally many custom weights "
                              "(%d) as outputs (%d).") %
                             (len(multioutput), n_outputs))
    y_type = 'continuous' if n_outputs == 1 else 'continuous-multioutput'

    return y_type, y_true, y_pred, multioutput

def mean_absolute_percentage_error(y_true, y_pred):
    y_type, y_true, y_pred, multioutput = _check_reg_targets(y_true, y_pred, multioutput='uniform_average')
    check_consistent_length(y_true, y_pred, None)
    output_errors = np.average(np.abs((y_true - y_pred) / y_true), axis=0)
    

    return (np.average(output_errors)*100)
```

##### 3. Model Assumptions

As Ridge Regression (included in sklearn.linear_model) is similar to Linear Regression. Most of the assumptions of Linear Regression (OLS Regression) will also be applicable to Ridge Regression.

From https://www.albert.io/blog/key-assumptions-of-ols-econometrics-review/, we can see some assumptions of OLS Regression.

1. Assumption-1 and Assumption-5 may hold true (upto a certain extent), as in the model we are supplying data for the past 7 days' price and also the rolling average calculation with a window of 10.

2. Assumption-2 should hold true, because Ridge Regression considers each data point (dependent variable) as unique, and only related to its corresponding independent values (in which we have added the past few days' data).

3. Assumption-3 should be true because it is natural for the model to sometimes predict higher values (than the actual) and sometimes lower values. So, the mean of errors should be close to zero.

4. Assumption-4 will hold true as there is only one independent variable.

5. Please refer to Point-1 above.

6. Assumption-6 may or may not hold true, but as said in the post this assumption is not compulsory for the validity of the model.
